```python
import itertools
import random
```

```python
random.seed(1)
num_days = 29
```

```python
with open('teams2.dat') as f:
    teams = f.read().splitlines()
```

```python
matchups = list(itertools.permutations(teams, r=2))
```

```python
def alreadyPlayed(matchup, day_matchup):
    try:
        home, away = list(zip(*day_matchup))
    except:
        return False
    if matchup[0] in home or matchup[1] in home or matchup[0] in away or matchup[1] in away:
        return True
    else:
        return False
    
def alreadyMatched(matchup, all_matchup):
    if matchup in all_matchup:
        return True
    else:
        return False
    
def consecutive(matchup, prev_matchup):
    prev_home, prev_away = list(zip(*prev_matchup))
    if matchup[0] in prev_home or matchup[1] in prev_away:
        return True
    else:
        return False
```

```python
def sortGames():
    season_matchup = []
    all_matchup = []
    day = 0
    while day < num_days:
        day_matchup = []
        for x in matchups:
            if not alreadyPlayed(x, day_matchup) and not alreadyMatched(x, all_matchup):
                try:
                    if not consecutive(x, season_matchup[day-1]):
                        day_matchup.append(x)
                except:
                    day_matchup.append(x)
                    
        #if len(day_matchup) < (len(teams)//2):
        #    random.shuffle(matchups)
        #    continue
        season_matchup.append(day_matchup)
        all_matchup += day_matchup
        random.shuffle(matchups)
        print(len(day_matchup))
        day+=1
    return season_matchup
```

```python
def printMatchups(season_matchup):
        for i in range(len(season_matchup)):
            home, away = list(zip(*season_matchup[i]))
            print('Round '+str(i))
            for j in range(len(home)):
                print(str(home[j])+' - '+str(away[j]))
            print('\n')
```

```python
season_matchup = sortGames()
```

```python
printMatchups(season_matchup)
```
